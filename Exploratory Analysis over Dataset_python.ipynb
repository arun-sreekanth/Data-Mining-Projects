# Task 2: Aggregation & Filtering & Rank (30 points)
In this task, we will perform some very high level aggregation and filtering operations. 
Then, we will apply ranking on the results for some tasks.

```python
# 5 points
# Task 2-a: Find out how many vehicles registered in the year 2018 which has fuel type 'diesel'
vehicles_registered = df_data[(df_data['yearOfRegistration'] == 2018) & (df_data['fuelType'] == 'diesel')].shape[0]
print(f"Number of vehicles with diesel type in the year 2018 registered are: {vehicles_registered}")

# 10 points
# Task 2-b: Print the records of vehicles whose price is in between 10,000 and 50,000 (excluding these prices) which was registered in the month of april(4)

print(f"Records of Vehicles whose price is in between 10000 and 50000(excluding those prices) which was registered in the month of April(4) are :\n")
records_of_vehicles = df_data[(10000 < df_data['price']) & 
                              (df_data['price'] < 50000) & 
                              (df_data['monthOfRegistration'] == 4)]
print(records_of_vehicles)


# 5 points
# Task 2-c: Discover the top 5 models with manuell gearbox and print a list of them.

model = df_data[df_data['gearbox']=='manuell']
grouped_data=model.groupby('model').size().reset_index(name ='count')
sorted_data = grouped_data.sort_values('count',ascending=False)
top_5_models = sorted_data.head(5)
print("Top 5 Models with Manuell gearbox:")
for model in top_5_models['model']:
    print(model)



# 10 points
# Task 2-d: Print records of vehicles which sold out with 'Gesuch' offertype with prices lower than 10,000

print(f"Records of vehicles which sold out with 'Gesuch' offertype with prices lower than 10,000 are :")
vehicles_gesuch = df_data.loc[(df_data['offerType'] == 'Gesuch') & (df_data['price'] < 10000), :]
print(vehicles_gesuch)


```

TASK 2-a : The .shape[] function in the first line of code is used to get the number of rows in the resulting dataframe which gives the count of vehicles registered in 2018 with diesel fuel type.
The second line of the code prints the count of the vehicles with diesel type registered in the year 2018 in a formatted string.

TASK 2-b : In this second task it filters the records based on the price given in the condition. These filtered records are stored in the records_of_vehicles variable printed using the print() function. The resulting output shows the filtered records, which should only include those vehicles meeting the specified criteria.

TASK 2-c : Firstly, a new dataframe is created called 'model' which only includes rows with a manual gearbox. Then, the 'model' dataframe is grouped by car model and the size of each group is counted and stored in a new column called 'count'. The resulting grouped data is then sorted in descending order based on the count of each car model. Finally, the top 5 car models with a manual gearbox are printed to the console.

TASK 2-d : Here we are printing the  records of vehicles which sold out with 'Gesuch' offertype with prices lower than 10,000. for this we are using &(and) based on the condition with offertype 'Gesuch' and price lower than 10,000 which will return values only if the condition is True.

```python
# 10 points
# Task 3-a: Display the boxplot indicating the distance travelled by each type of vehicle
sns.set(rc={'figure.figsize':(10,6)})
sns.boxplot(x='vehicleType',y='kilometer',data=df_data,palette = "colorblind")
plt.title("Distance Travelled by each type of the vehicle:")
plt.xlabel("vehicleType")
plt.ylabel("kilometer")
```

TASK 3-a : In the above code figsize sets the figure size of the plot.The palette parameter is used to set the color palette of the plot.Title sets the title of the plot. We plot using the boxplot with x labels as vehicleType and y lables as kilometer. The resulting boxplot visualizes the distribution of the distance traveled by each type of vehicle in the dataset, allowing us to compare their medians, quartiles, and outliers. It can provide insights into whether some types of vehicles tend to travel longer distances than others.

```python
# 20 points
# Task 3-b: Display a pie chart that represents brands and display percentages and names of brands in legend respectively
count_of_each_brand = df_data['brand'].value_counts()
plt.figure(figsize=(9,9))
plt.pie(count_of_each_brand.values, labels=count_of_each_brand.index, autopct='%1.1f%%')
plt.title('Brands and Percentage of Brands', fontsize=20)
#adding legend to show the various types of brands 
plt.legend(title="Name of the Brands", loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

TASK 3-b : In the code the value counts of each brand is calculated using the value_counts() method and stored in the variable count_of_each_brand.
Pie chart is created using the pie() function from matplotlib. The values and labels for the pie chart are taken from the count_of_each_brand variable, with the percentages shown using the autopct parameter.
The title of the pie chart is set using the title() function, with a fontsize of 20. A legend is also added to show the various types of brands using the legend() function, with the title set to "Name of the Brands" and positioned at the center left of the chart using the loc and bbox_to_anchor parameters.
Finally, the chart is displayed using the show() function.
